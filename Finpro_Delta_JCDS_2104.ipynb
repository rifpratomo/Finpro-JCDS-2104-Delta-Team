### **A. Latar Belakang**

Dalam era bisnis modern, di mana teknologi terus berkembang pesat, E-commerce telah menjadi salah satu pilar utama dalam perekonomian global. E-commerce, singkatan dari perdagangan elektronik, mencakup berbagai kegiatan bisnis yang dilakukan secara online, mulai dari pembelian dan penjualan produk hingga layanan konsumen. Sebagai sarana yang memungkinkan konsumen untuk berbelanja dan berinteraksi dengan perusahaan tanpa batasan geografis, E-commerce telah membuka pintu untuk peluang pertumbuhan yang signifikan.

Dalam konteks bisnis E-commerce modern, retensi pelanggan menjadi faktor krusial yang sangat memengaruhi kelangsungan perusahaan di tengah persaingan yang semakin ketat. Profitabilitas suatu perusahaan E-commerce tidak hanya ditentukan oleh kualitas produk atau layanan yang ditawarkan, tetapi juga oleh kemampuan perusahaan dalam mempertahankan dan memperluas basis pelanggan yang setia. Pengembangan metode pembelajaran yang dapat mencegah dan mengurangi customer churn (pelanggan yang beralih) menjadi esensial dalam strategi bisnis E-commerce.

Faktor-faktor yang menyebabkan *customer churn* (pelanggan beralih) dalam konteks bisnis E-commerce sangat beragam dan dapat dipengaruhi oleh berbagai elemen. Beberapa faktor utama yang dapat memicu pelanggan untuk meninggalkan suatu platform E-commerce melibatkan kombinasi dari aspek layanan, pengalaman pengguna, dan dinamika pasar. Berikut adalah beberapa faktor yang dapat memengaruhi *customer churn*:

1. **Kualitas Layanan Pelanggan:**
   - Kurangnya respons atau penanganan yang lambat terhadap pertanyaan dan masalah pelanggan dapat membuat mereka kecewa dan mencari alternatif yang lebih responsif.

2. **Pengalaman Pengguna yang Buruk:**
   - Antarmuka yang rumit, lambatnya waktu pemuatan halaman, atau kesulitan dalam navigasi dapat merugikan pengalaman pengguna, yang pada gilirannya dapat menjadi penyebab pelanggan meninggalkan platform.

3. **Harga dan Penawaran Bersaing:**
   - Penawaran harga yang tidak kompetitif atau kurangnya diskon dan promosi dapat membuat pelanggan mencari pilihan yang lebih ekonomis di tempat lain.

4. **Ketidakpuasan Produk atau Layanan:**
   - Produk atau layanan yang tidak sesuai dengan harapan pelanggan atau adanya masalah kualitas dapat menjadi alasan kuat untuk beralih ke penyedia yang lebih memenuhi kebutuhan mereka.

5. **Komunikasi yang Tidak Efektif:**
   - Kurangnya komunikasi yang efektif, termasuk kurangnya pemberitahuan tentang penawaran khusus atau pembaruan produk, dapat membuat pelanggan merasa diabaikan.

6. **Perubahan Preferensi Pelanggan:**
   - Perubahan kebutuhan atau preferensi pelanggan seiring waktu dapat membuat mereka mencari penyedia yang lebih cocok dengan perubahan tersebut.

7. **Persaingan yang Intensif:**
   - Keberadaan pesaing yang menawarkan layanan serupa dengan harga lebih baik atau dengan keunggulan lainnya dapat memotivasi pelanggan untuk mencari alternatif.

8. **Ketidakpuasan terhadap Kebijakan Bisnis:**
   - Kebijakan pengembalian yang tidak fleksibel, biaya tambahan yang tidak terduga, atau masalah terkait dengan kebijakan bisnis dapat menyebabkan ketidakpuasan dan *customer churn*.

Hal ini menunjukkan bahwa setiap bisnis, termasuk perusahaan E-commerce, harus menjaga hubungan dengan pelanggan yang setia. Evaluasi terhadap pelanggan yang meninggalkan layanan dalam jangka waktu tertentu, atau yang dikenal sebagai *customer churn*, merupakan langkah penting. Kehilangan pelanggan yang loyal dapat merugikan karena perusahaan harus mengeluarkan biaya hingga 16 kali lipat lebih besar untuk menggantikan pelanggan tersebut dengan pelanggan baru. Dengan memiliki wawasan tentang pelanggan mana yang berisiko untuk beralih, perusahaan dapat menyusun strategi pemasaran yang lebih efektif dan memberikan perhatian khusus kepada kelompok pelanggan ini."

### **E. Metrik Evaluasi**

Target:   
0 (Negatif) : Pelanggan tidak *churn*  
1 (Positif) : Pelanggan *churn*
- **True Positive (TP)**: Pelanggan diprediksi dan kenyataannya melakukan churn
- **False Positive (FP)**: Pelanggan diprediksi melakukan churn, namun pada kenyataannya customer tidak melakukan churn
- **False Negative (FN)**: Pelanggan diprediksi tidak melakukan churn, namun pada kenyataannya customer melakukan churn
- **True Negative (TN)**: Pelanggan diprediksi dan kenyataannya tidak melakukan churn

Dalam melakukan prediksi, kesalahan dapat terjadi, yaitu:
- **Type 1 Error** : False Positive
    - Konsekuensi : Perusahaan mengeluarkan biaya pada pelanggan yang kurang tepat.
- **Type 2 Error** : False Negative
    - Konsekuensi : Hilangnya pelanggan loyal

Model yang telah dibuat harus mampu mengurangi kerugian perusahaan akibat kehilangan pelanggan atau mengurangi kesalahan prediksi False Negatif. Namun, perlu juga dijaga agar perusahaan tidak menghabiskan biaya yang tidak diperlukan untuk memberikan promosi kepada pelanggan yang sebenarnya tidak berisiko (False Positive). Oleh karena itu, kita menggunakan metrik utama yaitu **F2 score**.

F2 score adalah metrik evaluasi klasifikasi yang memperhitungkan presisi dan recall. Nilai F2 score didasarkan pada parameter beta, di mana jika beta lebih besar dari 1, maka lebih menekankan recall daripada presisi. Metrik ini berguna untuk menilai kinerja model klasifikasi, terutama dalam kasus di mana kesalahan False Negatif lebih kritis daripada False Positive.

#### **Outliers**

Beberapa model nanti yang akan digunakan ketika *modeling* memiliki sifat yang sensitif terhadap outlier seperti Logistic Regression, K-Nearest Neighbour, dsb. Oleh karena itu perlu dilakukan analisis terhadap outlier.

```python
# Membuat variabel baru berisi kolom-kolom numerik kontinu
outliers_numerik = data[['Tenure', 'WarehouseToHome', 'HourSpendOnApp', 'OrderAmountHikeFromlastYear', 'CouponUsed', 'OrderCount', 'DaySinceLastOrder', 'CityTier', 'NumberOfDeviceRegistered', 'SatisfactionScore', 'NumberOfAddress', 'CashbackAmount']]

# Menampilkan Outliers dengan Boxplot
plt.figure(figsize=(20,20))
for x, kolom in enumerate(outliers_numerik, 1):
    plt.subplot(4, 3, x)  # Menggunakan x sebagai indeks subplot
    sns.boxplot(data=data[kolom], orient='horizontal') #visualisasi outliers dengan boxplot
    plt.title(f'Boxplot Kolom {kolom}') #judul boxplot
    plt.xlabel(kolom)

plt.tight_layout()
plt.show()
```

Dari masing-masing Variable terlihat jelas bahwasannya ada outlier. Hal ini bisa di identifikasi dari adanya titik titik yang berada jauh dari gambar boxplotnya. Dari plot diatas dapat kita simpulkan hanya variabel SatisfactionScore saja yang tidak ada outliernya. Setelah kita mengetahui variable mana saja yang terdapat Outlier, selanjutnya kita akan atasi Outlier dengan menggunakan metode interquartile range (IQR).

```python
# Menampilkan lower bound (quartil bawah), upper bound (quartil atas), outlier, dan persentase outlier
def outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)
    condition = (data[column] < lower_bound) | (data[column] > upper_bound)
    outliers = data[condition]
    return lower_bound, upper_bound, outliers

for column in outliers_numerik:
    lower_bound, upper_bound, data_outliers = outliers(data, column)
    data_outliers_count = len(data_outliers)
    persentase_outliers = round(data_outliers_count / len(data[column]) * 100, 2)

    print(f'Hasil pada kolom "{column}" yaitu, Lower bound = {round(lower_bound, 2)}, Upper bound = {round(upper_bound, 2)}, Total outliers = {data_outliers_count}, dan Persentase outliersnya = {persentase_outliers}')
```

Berdasarkan analisis sebelumnya, outlier dalam setiap kolom data tidak akan dihapus, karena dianggap sebagai nilai yang wajar apabila memiliki nilai variansi yang tinggi. Meskipun demikian, beberapa kolom akan dikelompokkan untuk mempermudah proses analisis data.

#### **4. Pengaruh Tenure Pelanggan Terhadap Churn**

```python
dataeda  = data.copy()
dataeda['Tenure']=dataeda['Tenure'].astype(int)
dataeda['Tenure']=dataeda['Tenure'].astype(object)

# Membuat fungsi untuk melakukan transformasi pada data
def transformasi(angka):
    if 13 <= angka <= 18:
        return '13-18 Month'
    elif 19 <= angka <= 24:
        return '19-24 Month'
    elif angka > 24:
        return 'Over 2 Year'
    else:
        return str(angka)

# Menerapkan transformasi pada kolom 'angka'
dataeda['transformed'] = dataeda['Tenure'].apply(transformasi)

# Hitung value counts pada kolom "Tenure"
value_counts_tenure = dataeda['transformed'].value_counts()

# Hitung value counts pada kolom "Tenure" jika Churn == 1
value_counts_churned_tenure = dataeda[dataeda['Churn'] == 1]['transformed'].value_counts()

# Hitung rasio
ratio_churned_tenure = value_counts_churned_tenure / value_counts_tenure

# Buat DataFrame hasil
result_df = pd.DataFrame({
    'Jumlah': value_counts_tenure,
    'Jumlah Churned': value_counts_churned_tenure,
    'Rasio Churned': ratio_churned_tenure
})

# Isi nilai NaN dengan 0 (jika ada)
result_df = result_df.fillna(0)

# Urutkan indeks sesuai dengan urutan yang diinginkan
desired_order =['Over 2 Year', '19-24 Month', '13-18 Month'] + [str(i) for i in range(12, -1, -1)]
result_df = result_df.reindex(desired_order)

# Format kolom 'Rasio Churned' menjadi dua angka desimal
result_df['Rasio Churned'] = result_df['Rasio Churned'].apply(lambda x: round(x, 4))
result_df['Jumlah Churned'] = result_df['Jumlah Churned'].astype(int)

```

```python
# Buat stacked proportional horizontal bar graph dengan label rasio persen
fig, ax = plt.subplots(figsize=(10, 10))

# Plot bar untuk 'Jumlah Churned'
bars1 = ax.barh(result_df.index, result_df['Jumlah Churned'], label='Churned', color='orange')

# Plot bar untuk 'Jumlah - Jumlah Churned'
bars2 = ax.barh(result_df.index, result_df['Jumlah'] - result_df['Jumlah Churned'], left=result_df['Jumlah Churned'], label='Total Customer per Tenure', color='blue')

# Tambahkan label dan judul
ax.set_ylabel('Tenure')
ax.set_xlabel('Jumlah')
ax.set_title('Rasio Churn Berdasarkan Tenure')


# Tambahkan label pada setiap bar
for bar1, bar2, label in zip(bars1, bars2, result_df['Rasio Churned']):
    xval = bar1.get_width() + 50
    ax.text(xval, bar1.get_y() + bar1.get_height() / 2, f'{label * 100:.2f}%', ha='center', va='center', color='white',fontweight='bold')

# Tambahkan label pada setiap bar
for bar1, bar2, churn, total in zip(bars1, bars2, result_df['Jumlah Churned'],result_df['Jumlah'] ):
    xval = bar1.get_width() + bar2.get_width() +60
    label = f'({churn}/{total})'
    ax.text(xval, bar1.get_y() + bar1.get_height() / 2, f'{label}', ha='center', va='center', color='black')


ax.legend()

# Sesuaikan batas-batas subplot untuk memperbesar kotak subplot tanpa mengubah ukuran total grafik
plt.xlim(0, 1100)

# Tampilkan grafik
plt.show()
```

Grafik tersebut menampilkan distribusi pelanggan berdasarkan durasi (tenure) mereka dan proporsi yang berhenti berlangganan, dengan pelanggan yang berhenti ditandai dengan warna kuning sementara yang tetap berlangganan ditampilkan dengan warna biru.

Tingginya tingkat churn pada periode awal (0-1 tenure) memperlihatkan ketidakmampuan dalam mempertahankan pelanggan pada fase fase awal penggunaan layanan. Namun, seiring berjalannya waktu, terlihat penurunan jumlah pelanggan yang berhenti, menggambarkan upaya yang berhasil dalam mempertahankan pelanggan yang telah menggunakan layanan untuk jangka waktu yang lebih lama.

## **Data Preprocessing & Feature Engineering**

Setelah memahami data serta melakukan data analisis, langkah-langkah yang dilakukan dalam proses preprocessing yaitu:
1. **Scaling**
    - Melakukan Robust Scaling terhadap variabel numerik agar antar variabel memiliki skala yang sama. Variable numerik ialah 'Tenure', 'WarehouseToHome', 'HourSpendOnApp', 'NumberOfDeviceRegistered', 'SatisfactionScore', 'NumberOfAddress', 'OrderAmountHikeFromlastYear', 'CouponUsed', 'OrderCount', 'DaySinceLastOrder', dan 'CashbackAmount'.

    - Melakukan OneHotEncoding pada kolom 'PreferredLoginDevice', 'PreferredPaymentMode', 'Gender', 'PreferedOrderCat', 'MaritalStatus', dan 'Complain' karena memiliki kurang dari 5 kategori dan tidak ada tingkatan dalam kategori.
    
Membuat transformer encoding sesuai dengan penjabaran diatas.

```python
# Skema transformer dengan scaling

scaler = Pipeline([
    ('Missing Value', IterativeImputer(max_iter=10)),
    ('Scaler', RobustScaler())
])

transformer1 = ColumnTransformer([
    ('OHE', OneHotEncoder(drop='first'),['PreferredLoginDevice', 'PreferredPaymentMode', 'Gender', 'PreferedOrderCat', 'MaritalStatus', 'Complain']),
    ('Scaler', scaler, ['Tenure', 'WarehouseToHome', 'HourSpendOnApp', 'NumberOfDeviceRegistered', 'SatisfactionScore', 'NumberOfAddress', 'OrderAmountHikeFromlastYear', 'CouponUsed', 'OrderCount', 'DaySinceLastOrder', 'CashbackAmount'])
], remainder='passthrough')

transformer1
```

```python
# Skema transformer tanpa scaling

# Menginisialisasi IterativeImputer
imputer = IterativeImputer(max_iter=10)

transformer2 = ColumnTransformer([
    ('OHE', OneHotEncoder(drop='first'),['PreferredLoginDevice', 'PreferredPaymentMode', 'Gender', 'PreferedOrderCat', 'MaritalStatus', 'Complain']),
    ('Missing Value', imputer,['Tenure', 'WarehouseToHome', 'HourSpendOnApp', 'OrderAmountHikeFromlastYear', 'CouponUsed', 'OrderCount', 'DaySinceLastOrder', 'CashbackAmount'])
], remainder='passthrough')

transformer2
```

## **Model Benchmarking**
---

Terdapat beberapa metode klasifikasi yang akan diuji coba sebagai penentuan model terbaik untuk dataset ini, diantaranya ialah Logistic Regression, KNearest Neighbour, Decision Tree, Random Forest, XGBoost, Voting, dan Stacking.

Beberapa hal yang perlu dilakukan dalam model benchmarking adalah:
- Dilakukan resampling karena terdapat imbalance data berdasarkan hasil data analisis. Metode resampling yang digunakan adalah RandomOverSampler untuk meningkatkan kelas minoritas mengikuti kelas mayoritas, serta RandomUnderSampling dan Near Miss yang berfungsi menurunkan kelas mayoritas mengikuti kelas minoritas.
- Scaler yang digunakan adalah Robust Scaler, karena scaler ini tidak sensitif terhadap outlier.
- Metrik yang digunakan adalah F2 Score. F2 Score lebih baik digunakan karena fokus pada menghindari False Negatives, yang penting dalam mengidentifikasi pelanggan yang berpotensi *churn*.
- Dilakukan 4 kondisi benchmarking model untuk mengetahui model yang terbaik, yaitu:
    1. Modeling Menggunakan Pipeline Dengan Kondisi Scaling dan Resampling
    2. Modeling Menggunakan Pipeline Dengan Kondisi Scaling Tetapi Tidak Resampling
    3. Modeling Menggunakan Pipeline Dengan Kondisi Tidak Scaling Tetapi Resampling
    4. Modeling Menggunakan Pipeline Dengan Kondisi Tidak Scaling dan Tidak Resampling
    

#### **1. Modeling Menggunakan Pipeline Dengan Kondisi Scaling dan Resampling**

```python
# Skema 1 : Modeling dengan Encoding, Scaling, dan Resampling

#Algoritma model yang digunakan
logreg = LogisticRegression(random_state=2023)
dt = DecisionTreeClassifier(random_state=2023)
knn = KNeighborsClassifier()
svm = SVC(random_state=2023)
xgb = XGBClassifier()
rf = RandomForestClassifier(random_state=2023)

vc_hard = VotingClassifier(estimators=[
    ('model1', logreg),
    ('model2', dt),
    ('model3', knn),
    ('model4', svm),
    ('model5', xgb),
    ('model6', rf),
],voting='hard')

sc_logreg = StackingClassifier(estimators=[
    ('model1', logreg),
    ('model2', dt),
    ('model3', knn),
    ('model4', svm),
    ('model5', xgb),
    ('model6', rf),
],final_estimator=LogisticRegression(random_state=2023))

sc_dt = StackingClassifier(estimators=[
    ('model1', logreg),
    ('model2', dt),
    ('model3', knn),
    ('model4', svm),
    ('model5', xgb),
    ('model6', rf),
],final_estimator=DecisionTreeClassifier(random_state=2023))

sc_knn = StackingClassifier(estimators=[
    ('model1', logreg),
    ('model2', dt),
    ('model3', knn),
    ('model4', svm),
    ('model5', xgb),
    ('model6', rf),
],final_estimator=KNeighborsClassifier())

sc_svm = StackingClassifier(estimators=[
    ('model1', logreg),
    ('model2', dt),
    ('model3', knn),
    ('model4', svm),
    ('model5', xgb),
    ('model6', rf),
],final_estimator=SVC(random_state=2023))

# Metode resampling yang digunakan
ros = RandomOverSampler()
nm = NearMiss()
rus = RandomUnderSampler()

model = [logreg,dt,knn,svm,xgb,rf,vc_hard,sc_logreg,sc_dt,sc_knn,sc_svm] # list algoritma model yang dipakai
resampling = [ros,nm,rus] # list metode resampling yang digunakan
f2_score = [] # list kosong untuk menampung hasil f2 score
std = [] # list kosong untuk menampung hasil standar deviasi

# Menampilkan Looping setiap model dengan resampling menggunakan skema transformer1
for i in model :
    for j in resampling :
        model_pipe = Pipeline([
            ('Preprocess',transformer1),
            ('imba',j),
            ('clf',i)
        ])
        pipe_cv = cross_val_score(model_pipe, xtrain, ytrain, cv=5, scoring=make_scorer(fbeta_score, beta=2)) #cross validation
        f2_score.append(pipe_cv.mean()) #rata-rata f2 score dari cross validation
        std.append(pipe_cv.std()) #nilai standar deviasi dari cross validation

# Memasukkan hasil modeling sesuai dengan urutan f2 score terbesar kedalam data frame
pd.DataFrame({
    'Model' : ['Logistic Regression ROS','Logistic Regression Near Miss','Logistic Regression RUS',
               'DT ROS','DT Near Miss','DT RUS',
               'KNN ROS','KNN Near Miss','KNN RUS',
               'SVM ROS','SVM Near Miss','SVM RUS',
               'XGBoost ROS','XGBoost Near Miss','XGBoost RUS',
               'Random Forest ROS','Random Forest Near Miss','Random Forest RUS',
               'Hard Voting ROS','Hard Voting Near Miss','Hard Voting RUS',
               'Stacking Classifier Logreg ROS','Stacking Classifier Logreg Near Miss','Stacking Classifier Logreg RUS',
               'Stacking Classifier DT ROS','Stacking Classifier DT Near Miss','Stacking Classifier DT RUS',
               'Stacking Classifier KNN ROS','Stacking Classifier KNN Near Miss','Stacking Classifier KNN RUS',
               'Stacking Classifier SVM ROS','Stacking Classifier SVM Near Miss','Stacking Classifier SVM RUS',
               ],
    'F2 Score Mean' : f2_score,
    'F2 Score Std' : std
}).sort_values('F2 Score Mean', ascending=False)

```

```python
# Skema 3 : Modeling dengan Encoding dan Resampling

#Algoritma model yang digunakan
logreg = LogisticRegression(random_state=2023)
dt = DecisionTreeClassifier(random_state=2023)
knn = KNeighborsClassifier()
svm = SVC(random_state=2023)
xgb = XGBClassifier()
rf = RandomForestClassifier(random_state=2023)

vc_hard = VotingClassifier(estimators=[
    ('model1', logreg),
    ('model2', dt),
    ('model3', knn),
    ('model4', svm),
    ('model5', xgb),
    ('model6', rf),
],voting='hard')

sc_logreg = StackingClassifier(estimators=[
    ('model1', logreg),
    ('model2', dt),
    ('model3', knn),
    ('model4', svm),
    ('model5', xgb),
    ('model6', rf),
],final_estimator=LogisticRegression(random_state=2023))

sc_dt = StackingClassifier(estimators=[
    ('model1', logreg),
    ('model2', dt),
    ('model3', knn),
    ('model4', svm),
    ('model5', xgb),
    ('model6', rf),
],final_estimator=DecisionTreeClassifier(random_state=2023))

sc_knn = StackingClassifier(estimators=[
    ('model1', logreg),
    ('model2', dt),
    ('model3', knn),
    ('model4', svm),
    ('model5', xgb),
    ('model6', rf),
],final_estimator=KNeighborsClassifier())

sc_svm = StackingClassifier(estimators=[
    ('model1', logreg),
    ('model2', dt),
    ('model3', knn),
    ('model4', svm),
    ('model5', xgb),
    ('model6', rf),
],final_estimator=SVC(random_state=2023))

# Metode resampling yang digunakan
ros = RandomOverSampler()
nm = NearMiss()
rus = RandomUnderSampler()

model = [logreg,dt,knn,svm,xgb,rf,vc_hard,sc_logreg,sc_dt,sc_knn,sc_svm] # list algoritma model yang dipakai
resampling = [ros,nm,rus] # list metode resampling yang digunakan
f2_score = [] # list kosong untuk menampung hasil f2 score
std = [] # list kosong untuk menampung hasil standar deviasi

# Menampilkan Looping setiap model dengan resampling menggunakan skema transformer1
for i in model :
    for j in resampling :
        model_pipe = Pipeline([
            ('Preprocess',transformer2),
            ('imba',j),
            ('clf',i)
        ])
        pipe_cv = cross_val_score(model_pipe, xtrain, ytrain, cv=5, scoring=make_scorer(fbeta_score, beta=2)) #cross validation
        f2_score.append(pipe_cv.mean()) #rata-rata f2 score dari cross validation
        std.append(pipe_cv.std()) #nilai standar deviasi dari cross validation

# Memasukkan hasil modeling sesuai dengan urutan f2 score terbesar kedalam data frame
pd.DataFrame({
    'Model' : ['Logistic Regression ROS','Logistic Regression Near Miss','Logistic Regression RUS',
               'DT ROS','DT Near Miss','DT RUS',
               'KNN ROS','KNN Near Miss','KNN RUS',
               'SVM ROS','SVM Near Miss','SVM RUS',
               'XGBoost ROS','XGBoost Near Miss','XGBoost RUS',
               'Random Forest ROS','Random Forest Near Miss','Random Forest RUS',
               'Hard Voting ROS','Hard Voting Near Miss','Hard Voting RUS',
               'Stacking Classifier Logreg ROS','Stacking Classifier Logreg Near Miss','Stacking Classifier Logreg RUS',
               'Stacking Classifier DT ROS','Stacking Classifier DT Near Miss','Stacking Classifier DT RUS',
               'Stacking Classifier KNN ROS','Stacking Classifier KNN Near Miss','Stacking Classifier KNN RUS',
               'Stacking Classifier SVM ROS','Stacking Classifier SVM Near Miss','Stacking Classifier SVM RUS',
               ],
    'F2 Score Mean' : f2_score,
    'F2 Score Std' : std
}).sort_values('F2 Score Mean', ascending=False)

```
